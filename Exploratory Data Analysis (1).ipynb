```python
# Find all the missing values in the dataset. If the missing values of a particular column are more than 80%, drop.
# If the missing values are less than 80%, handle them appropriately

missing_values = df.isnull().sum().sort_values(ascending = False)
missing_values
```

```python
# percentage of missing values in the dataset.
percent = round((missing_values / len(df)) * 100)
print(percent)
```

```python
# dropping columns that has more than 80% missing values.
columns_to_drop = percent[percent > 80].index
df = df.drop(columns=columns_to_drop)
```

```python
df.describe()
```

```python
# handling missing values that are less than 80% in the dataset.
# for numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())
```

```python
df.info()
```

```python
# filling missing values for categorical columns
categorical_columns = df.select_dtypes(include='object').columns
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])
```

```python
df.info()
```

```python
# finding outliers in the dataset
for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

print(df[(df[col] < lower_limit) | (df[col] > upper_limit)])
```

```python
df = df[(df[col] >= lower_limit) & (df[col] <= upper_limit)]
print(df)
```

# Part 2: User Overview Analysis


```python
# the column with unique values in the dataset. 
unique_imei = df['IMEI'].nunique()
print(unique_imei)

unique_imsi = df['IMSI'].nunique()
print(unique_imsi)

unique_msisdn = df['MSISDN/Number'].nunique()
print(unique_msisdn)
```

```python
# Aggregate the number of xDR sessions, the session duration, the total download (DL) and upload (UL) data, the total data volume (in Bytes) during this session for each application

data = df.groupby('IMSI').agg({
    'Bearer Id': 'count',
    'Dur. (ms)': 'sum',
    'Total UL (Bytes)': 'sum',
    'Total DL (Bytes)': 'sum',
    'Social Media DL (Bytes)': 'sum',
    'Social Media UL (Bytes)': 'sum',
    'Google DL (Bytes)': 'sum',
    'Google UL (Bytes)': 'sum',
    'Email DL (Bytes)': 'sum',
    'Email UL (Bytes)': 'sum',
    'Youtube DL (Bytes)': 'sum',
    'Youtube UL (Bytes)': 'sum',
    'Netflix DL (Bytes)': 'sum',
    'Netflix UL (Bytes)': 'sum',
    'Gaming DL (Bytes)': 'sum',
    'Gaming UL (Bytes)': 'sum',
    'Other DL (Bytes)': 'sum',
    'Other UL (Bytes)': 'sum',
    })
```

```python
# The 3 most used applications by customers
application_data = {
    'Social media': data['Social Media UL (Bytes)'].sum() + data['Social Media DL (Bytes)'].sum(),
    'Google': data['Google UL (Bytes)'].sum() + data['Google DL (Bytes)'].sum(),
    'Email': data['Email UL (Bytes)'].sum() + data['Email DL (Bytes)'].sum(),
    'Youtube': data['Youtube UL (Bytes)'].sum() + data['Youtube DL (Bytes)'].sum(),
    'Netflix': data['Netflix UL (Bytes)'].sum() + data['Netflix DL (Bytes)'].sum(),
    'Gaming': data['Gaming UL (Bytes)'].sum() + data['Gaming DL (Bytes)'].sum(),
    'Other': data['Other UL (Bytes)'].sum() + data['Other DL (Bytes)'].sum()
      }


application_data
```

```python
application_data_df = pd.DataFrame(list(application_data.items()), columns=['Applications' , 'Total app data (Bytes)'])
application_data_df
```

```python
grouped_data = application_data_df.sort_values(by = 'Total app data (Bytes)' ,ascending=False)
top_3_applications = grouped_data.head(3)
top_3_applications
```

```python
# plotting a bar chart of the top three most used applications by users.

plt.bar(top_3_applications['Applications'], top_3_applications['Total app data (Bytes)'])
```

# Part 5: Feature Engineering


```python
# Create a new column for data usage ratios for Google, email, YouTube, Netflix, gaming, social media, and other. 

df['Google Data Ratio'] = df['Google UL (Bytes)'] / df['Google DL (Bytes)']
df['Email Data Ratio'] = df['Email UL (Bytes)'] / df['Email DL (Bytes)']
df['Youtube Data Ratio'] = df['Youtube UL (Bytes)'] / df['Youtube DL (Bytes)']
df['Netflix Data Ratio'] = df['Netflix UL (Bytes)'] / df['Netflix DL (Bytes)']
df['Gaming Data Ratio'] = df['Gaming UL (Bytes)'] / df['Gaming DL (Bytes)']
df['Social Media Data Ratio'] = df['Social Media UL (Bytes)'] / df['Social Media DL (Bytes)']
df['Other Data Ratio'] = df['Other UL (Bytes)'] / df['Other DL (Bytes)']
```

```python
df.head()
```

```python
# Create a new column 'Total Data' that sums the 'Total DL (Bytes)' and 'Total UL (Bytes)
df['Total Data'] = df['Total DL (Bytes)'] + df['Total UL (Bytes)']
df.head()
```
